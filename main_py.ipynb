```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

store_data=[]
urls = [
    "https://store.starbucks.co.jp/detail-194/",
    "https://store.starbucks.co.jp/detail-1358/",
    "https://store.starbucks.co.jp/detail-4265/"
    ]
headers = {'User-Agent': 'Mozilla/5.0'}

for url in urls:
  response = requests.get(url, headers=headers)
  response.encoding = response.apparent_encoding
  soup = BeautifulSoup(response.text, 'html.parser')
  # ここで文字コードを明示的にUTF-8にする
  response.encoding = 'utf-8'


  # 要素を抽出
  name_element = soup.find('div', class_='store-detail__title-text')
  address_element = soup.find('div', class_='store-detail__text-detail line-height-22')
  # 電話番号と営業時間は同じクラス名なので find_all でリストとして取得
  details = soup.find_all('div', class_="store-detail__text-detail Sodo-text number-text")

  # 辞書として作成
  store_info = {
        '店名': name_element.text.strip() if name_element else 'なし',
        '住所': address_element.text.strip() if address_element else 'なし',
        # リストの順番で指定（0番目：電話、1番目：営業時間）
        '電話番号': details[0].text.strip() if len(details) > 0 else 'なし',
        '営業時間': details[1].text.strip() if len(details) > 1 else 'なし',
    }

  store_data.append(store_info)
  print(f"取得完了: {url}")
  time.sleep(1) # サーバーへのマナーとして1秒待機

# 表形式に変換して表示
df = pd.DataFrame(store_data)
print(df)

# CSVに保存（Excelで開くときに文字化けしない設定）
df.to_csv("starbucks_list.csv", index=False, encoding='utf-8-sig')
print("\nstarbucks_list.csv に保存しました！")
```
